## Imputing categorical integer values

```python
def impute_cat_cols(df):
    cat_cols = ['Sex', 'Smoking', 'BPTreatment', 'PrevalentDiabetes', 'PrevalentCHD', 'PrevalentHFAIL']

    for cols in cat_cols:
        mask = df[cols].isnull() 
        original = df[cols]
        residence_encoder = LabelEncoder()
        df[cols] = residence_encoder.fit_transform(df[cols])

        #if there is NaN values
        if mask.any():
            #adding NaN again
            df[cols] = df[cols].where(~mask, original) 

            imputer = KNNImputer(n_neighbors=3)
            df[cols]= imputer.fit_transform(df[cols].values.reshape(-1, 1))

        df[cols] = pd.Categorical(df[cols].astype(int))
 
```

## Imputing numeric  values

```python
def impute_num_cols(df):
    num_cols = ['Age', 'BodyMassIndex', 'SystolicBP', 'NonHDLcholesterol']

    for cols in num_cols:
        if df[cols].isnull().any():
            imputer = SimpleImputer(missing_values=np.nan, strategy='median')
            df[cols] = imputer.fit_transform(df[cols].values.reshape(-1, 1))
```
